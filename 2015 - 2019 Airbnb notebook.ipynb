```python
# import all the necessary libraries for the analysis
import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
sns.set(font_scale = 1.5)

random.seed(23)
```

```python
# read in base listing file
nyc = pd.read_csv('./listings2019.csv', usecols = ['id','neighbourhood_group', 'neighbourhood', 'room_type',
                                                              'price','minimum_nights','number_of_reviews'])
nyc.rename(columns ={'id': 'listing_id'}, inplace = True)

# create empty data frames
a=pd.DataFrame(); b=pd.DataFrame(); c=pd.DataFrame(); d=pd.DataFrame(); df= pd.DataFrame()

#create a list of the first 4 empty data frames
s = [a,b,c,d]

# create a list of file names
ls = ['./calendar2018_2019.csv', './calendar2017.csv', './calendar2016.csv', './calendar2015.csv']


# read in each calender and merge it to the base listing data
for i in range(len(s)):

    x = pd.read_csv(ls[i])
    s[i] = pd.merge(nyc, x, on = 'listing_id')

    #delete x to save memory after each merge
    del x

# delete the base listings after reading and merging all data frames
del nyc,a,b,c,d,ls
    
# concat all the merged dataframes
df = pd.concat(s, ignore_index=True)

# clear memory
del s

# display top 5 rows of final data frame
df.head()
```

```python
# check proportion of na values in the price_x column
df['price_x'].isnull().mean()
```

```python
# check proportion of na values in the price_y column
df['price_y'].isnull().mean()
```

```python
# drop the price_y column since almost 70% is missing
df = df.drop(columns=['price_y'])

# check for more na values
df.isnull().any()
```

```python
# season function
def seasons(month):
    if month in ('January', 'February', 'March'):
        return 'Winter'
    elif month in ('April', 'May', 'June'):
        return 'Spring'
    elif month in ('July', 'August', 'September'):
        return 'Summer'
    else:
        return 'Fall'
```

```python
# change the price_x column to price
df.rename(columns ={'price_x': 'price'}, inplace = True)

# convert the available column to a binary column called booked
df['booked'] = np.where(df['available']=='f', 1, 0)

# create new month and year columns from the date column
df['month'] = pd.DatetimeIndex(df['date']).month_name()
df['year'] = pd.DatetimeIndex(df['date']).year.astype(str)

# create season variable
df['season'] = df['month'].apply(seasons)

df.head()
```

## Question 1:
### How has the number of bookings changed over the last 5 years?

```python
# How has the number of booking changed over the last 5 years?
booking_by_year = df.groupby('year').booked.sum()

# plot the trend line 
plt.xlabel("Price", fontsize = 18)
plt.ylabel("Number of Bookings", fontsize = 18)
booking_by_year.plot(kind='line', figsize =(10,5), marker = 'o');
plt.title('Booking Trend from 2015 to 2019 in New York City', fontsize=25);
```

## Question 5:
### How does neighborhood and price affect bookings?

```python
# How is the number of reservations affected by neighborhood, and price
price_neigh = df.groupby(['price','neighbourhood']).booked.sum().sort_values(ascending=False).head(10).reset_index()

price_neigh
```

```python
# create plot
plt.figure(figsize = (18, 12))
ax = sns.barplot(x ="price", y = "booked", hue= 'neighbourhood', data=price_neigh)
plt.title('Top 10 Neighbourhoods by Number of Bookings with Respect to Price', fontsize= 20)
plt.xlabel("Price", fontsize = 18)
plt.ylabel("Number of Bookings", fontsize = 18)

# change bar width code was sourced from stackoverflow
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)

change_width(ax, .35)

plt.show()
```

## Question 6:
### What are the top 10 earnings in each borough?

```python
# How is the number of reservations affected by neighborhood, and price
listin_earnings = df.groupby(['neighbourhood_group', 'neighbourhood','room_type',
                              'listing_id','price']).booked.sum().reset_index().sort_values(by=['booked',
                                                                                                'price'],ascending=False)
listin_earnings['5yr_earnings'] = np.multiply(listin_earnings['price'],listin_earnings['booked'])
```

```python
# Function to plot all bar charts
def box_plot(data, s):
    plt.figure(figsize = (16, 8))
    ax = sns.barplot(x ="Listing_Price", y = "5yr_earnings", hue= 'neighbourhood', data=data)
    plt.title('Top 10 ' + s + ' Earnings From 2015 - 2019', fontsize = 25)
    plt.xlabel('Listing Price', fontsize = 20)
    plt.ylabel('Earnings', fontsize = 20)
    
    # change bar width
    def change_width(ax, new_value) :
        for patch in ax.patches :
            current_width = patch.get_width()
            diff = current_width - new_value

            # we change the bar width
            patch.set_width(new_value)

            # we recenter the bar
            patch.set_x(patch.get_x() + diff * .5)

    change_width(ax, .35)
    
    return plt.show()
```

```python
# fuction to create an encoded listing ID for the x axis
def new_col(data):
    alp = ['a','b','c','d','e','f','g','h','i','j']
    x = []
    for i in range(len(alp)):
        x.append(alp[i].upper()+ ': $'+data.loc[i,'price'])
    return x
```

```python
# Bronx earnings
Bx_earnings = listin_earnings[listin_earnings['neighbourhood_group']=='Bronx'
                             ].head(10).reset_index().sort_values(by=['5yr_earnings'],ascending=False)
Bx_earnings['price'] = Bx_earnings['price'].astype(str)
Bx_earnings["Listing_Price"] = new_col(Bx_earnings)
box_plot(Bx_earnings, 'Bronx')
```
