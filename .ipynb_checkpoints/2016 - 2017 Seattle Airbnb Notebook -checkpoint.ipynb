```python
# import all the necessary libraries for the analysis
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
sns.set()
```

```python
# read in and display the top rows of the Seattle Airbnb listings datasets
sea_listings = pd.read_csv('./listings_seattle.csv', usecols = ['id', 'neighbourhood', 'room_type',
                                                                'price','minimum_nights','number_of_reviews'])
sea_listings.head()
```

```python
# read in and display the top rows of the Seattle Airbnb Calendar datasets
sea_calendar = pd.read_csv('./calendar_seattle.csv', usecols=['listing_id', 'date','available'])
sea_calendar.head()
```

```python
# rename column
sea_listings.rename(columns ={'id': 'listing_id'}, inplace = True)

# merge both data frames
df = pd.merge(sea_listings, sea_calendar, on = 'listing_id')

# convert the available column to a binary column called booked
df['booked'] = np.where(df['available']=='f', 1, 0)

# create new month and year columns from the date column
df['month'] = pd.DatetimeIndex(df['date']).month_name()
df['year'] = pd.DatetimeIndex(df['date']).year.astype(str)
```

```python
# How is the number of reservations affected by neighborhood, and price
listin_earnings = df.groupby(['neighbourhood','room_type',
                              'listing_id','price']).booked.sum().reset_index().sort_values(by=['booked',
                                                                                                'price'],ascending=False)
listin_earnings['price'] = listin_earnings['price'].str.replace(',', '').str.replace('$', '').astype(float)

listin_earnings['2yr_earnings'] = np.multiply(listin_earnings['price'],listin_earnings['booked'])
listin_earnings = listin_earnings.sort_values(by=['2yr_earnings'], ascending=False)
```

```python
def new_col(data):
    alp = ['a','b','c','d','e','f','g','h','i','j']
    x = []
    for i in range(len(alp)):
        x.append(alp[i].upper()+ ': $'+data.loc[i,'price'])
    return x
```

```python
def bar_plot(data, room):
    sns.set()
    plt.figure(figsize = (12, 8))
    ax = sns.barplot(x = 'Listing_Price', y = "2yr_earnings", hue="neighbourhood", data=data)
    plt.title('Top 10 ' + room + ' Earnings From 2016 -2017', fontsize = 25)
    plt.xlabel('Listing Price', fontsize = 20)
    plt.ylabel('Earnings', fontsize = 20)
    
    # change bar width
    def change_width(ax, new_value) :
        for patch in ax.patches :
            current_width = patch.get_width()
            diff = current_width - new_value

            # we change the bar width
            patch.set_width(new_value)

            # we recenter the bar
            patch.set_x(patch.get_x() + diff * .5)

    change_width(ax, .35)
    return plt.show()
```

```python
# Top 10 shared room listing earners
sh_earnings = listin_earnings[listin_earnings['room_type']== 'Shared room'].head(10).reset_index()
sh_earnings['price'] = sh_earnings['price'].astype(str)
sh_earnings["Listing_Price"] = new_col(sh_earnings)
```

```python
# Create plot
bar_plot(sh_earnings, 'Shared Room')
```

```python
# Top 10  Private room listing earners
pr_earnings = listin_earnings[listin_earnings['room_type']== 'Private room'].head(10).reset_index()
pr_earnings['price'] = pr_earnings['price'].astype(str)
pr_earnings["Listing_Price"] = new_col(pr_earnings)
```

```python
# Create plot
bar_plot(pr_earnings, 'Private Room')
```

```python
# Top 10 Entire home/apt listing earners
en_earnings = listin_earnings[listin_earnings['room_type']== 'Entire home/apt'].head(10).reset_index()
en_earnings['price'] = en_earnings['price'].astype(str)
en_earnings["Listing_Price"] = new_col(en_earnings)
```

```python
# Create plot
bar_plot(en_earnings, 'Entire Home/Apt')
```
